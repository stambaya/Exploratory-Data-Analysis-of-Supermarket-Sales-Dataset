```python
import pandas as pd  # data analysis
import seaborn as sns  # data visualization
import matplotlib.pyplot as plt  # data visualization
```

```python
df=pd.read_csv('Finance_data.csv')
```

```python
print(df)
```

```python
# Understanding the Data

print("Shape:", df.shape)
df.info()
df.describe()
df.isnull().sum()  # check for missing values
```

```python
# Gender Distribution

gender_count = df['gender'].value_counts()
sns.barplot(x=gender_count.index, y=gender_count.values)
plt.title("Gender Distribution")
plt.ylabel("Count")
plt.xlabel("Gender")
plt.show()
```

```python
#  Investment Avenues vs Gender

sns.countplot(data=df, x='Investment_Avenues', hue='gender')
plt.title("Investment Decision by Gender")
plt.xlabel("Has Investment?")
plt.ylabel("Count")
plt.show()
```

```python
# Purpose of Saving

purpose_count = df['What are your savings objectives?'].value_counts()
sns.barplot(y=purpose_count.index, x=purpose_count.values)
plt.title("Savings Objectives")
plt.xlabel("Count")
plt.ylabel("Objective")
plt.show()
```

```python
# Risk Factor (Expect %) Distribution

sns.countplot(data=df, x='Expect', hue='gender')
plt.title("Expected Return Range by Gender")
plt.xlabel("Expected Return")
plt.ylabel("Count")
plt.show()
```

```python
# Barplot: Most Preferred Investment Channel

columns = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
avg_scores = df[columns].mean().sort_values(ascending=False)

sns.barplot(x=avg_scores.index, y=avg_scores.values)
plt.xticks(rotation=45)
plt.title("Average Preference Score for Investment Avenues")
plt.ylabel("Average Score")
plt.xlabel("Investment Type")
plt.show()
```

```python
# Encoding if Needed

df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
df['Investment_Avenues'] = df['Investment_Avenues'].map({'Yes': 1, 'No': 0})
```
